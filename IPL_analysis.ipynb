# Match based analysis

# How many matches were played in total?

```python
matches=df.groupby("id").size().sum()
print(matches)
```

# Which team has won the most matches?

```python
winner=df.groupby("winner",as_index=False)["winner"].value_counts().sort_values("count",ascending=False)
ax=sns.barplot(x="winner",y="count",data=winner,palette="pastel")
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Winners")
plt.xticks(rotation=90)
plt.show()
```

# What are the top venues where the matches were held?

```python
venue=df.groupby("venue",as_index=False)["venue"].size().sort_values("size",ascending=False).head(10)
lab=sns.barplot(x="size",y="venue",data=venue,palette="plasma_r",alpha=0.7,orient="h")
for bar in lab.containers:
    lab.bar_label(bar)
plt.xticks(rotation=90)
plt.xlabel("Stadium")
plt.ylabel("Number of matches")
plt.title("Top Stadiums")
plt.show()
```

# What is the win percentage of each team?

```python
team_stat=df.groupby("winner",as_index=False)["winner"].value_counts().sort_values("count",ascending=False)
team_stat["team_perc"]=(team_stat["count"]*100)/matches
plt.pie(team_stat["team_perc"],autopct="%0.2f%%",radius=1.7,labels=team_stat["winner"])
plt.show()
```

# How does the performance of teams vary by venue?

```python
perfor=df.groupby(["winner","venue"],as_index=False)["winner"].value_counts().sort_values("winner",ascending=False)

pivot_table=perfor.pivot(index="venue",columns="winner",values="count").fillna(0).astype(int)
pivot_table
```

```python
ax=sns.heatmap(data=pivot_table,annot=True,cmap="BrBG_r",cbar_kws={"label": "Wins"},annot_kws={"fontsize":6})
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
plt.show()
```

# Which teams have the best head-to-head records?

```python
h2h=df.groupby(["team1","team2","winner"],as_index=False)["winner"].value_counts().sort_values("count",ascending=False).head(7)
h2h["game"]=df["team1"]+" VS "+df["team2"]
h2h
```

```python
lab=sns.barplot(x="game",y="count",data=h2h,palette="hsv",alpha=0.5)
for bar in lab.containers:
    lab.bar_label(bar)
plt.xticks(rotation=90)
plt.title("Best Head-to-Head Records",fontsize=12)
plt.xlabel("Match teams")
plt.ylabel("Number of wins")
plt.show()
```

```python
data=pd.read_csv("deliveries.csv")
data
```

```python
data.isnull().sum()
```

```python
data.drop(columns=["extras_type","player_dismissed","dismissal_kind","fielder"],inplace=True)
```

```python
data.head(3)
```

# Which players have the highest total runs?

```python
highest_runs=data.groupby("batter",as_index=False)["total_runs"].sum().sort_values("total_runs",ascending=False).head(10)
plt.figure(figsize=(12,6))
lab=sns.barplot(x="batter",y="total_runs",data=highest_runs,palette="autumn_r",alpha=0.7)
for bar in lab.containers:
    lab.bar_label(bar)
plt.title("Highest total runs",fontsize=12)
plt.xlabel("Batsmen")
plt.ylabel("Number of runs")
plt.show()
```

# Who are the most consistent batsmen (highest average)?


```python
add=data.groupby("batter",as_index=False)["total_runs"].sum()
count=data.groupby("batter",as_index=False)["total_runs"].count()
count.rename(columns={"total_runs":"innings"},inplace=True)
merged=add.merge(count,on="batter")
merged["average"]=merged["total_runs"]/merged["innings"]
merged["average"].astype(int)
consistent = merged.sort_values("average", ascending=False).head(10)
consistent
```

```python
plt.figure(figsize=(15,6))
lab=sns.barplot(x="batter",y="average",data=consistent,palette="cool_r",alpha=0.8)
for bar in lab.containers:
    lab.bar_label(bar)
plt.title("Most Consistent Player",fontsize=13)
plt.xlabel("Batsmen")
plt.ylabel("Average")
plt.show()
```

# What is the distribution of runs scored in boundaries (4s and 6s) versus singles?

```python
runs=data.groupby("batsman_runs",as_index=False)["batsman_runs"].count().reset_index().sort_values("index")
plt.figure(figsize=(8,6))
lab=sns.barplot(x="index",y="batsman_runs",data=runs,palette="BrBG",alpha=0.8)
for bar in lab.containers:
    lab.bar_label(bar)
plt.title("Boundaries VS Singles",fontsize=13)
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.show()
```

# Which bowlers have taken the most wickets?

```python
wickets=data.groupby("bowler",as_index=False)["is_wicket"].sum().sort_values("is_wicket",ascending=False).head(10)
plt.figure(figsize=(10,5))
lab=sns.barplot(x="bowler",y="is_wicket",data=wickets,palette="viridis",alpha=0.5)
for bar in lab.containers:
    lab.bar_label(bar)
plt.title("Most wickets",fontsize=13)
plt.xlabel("Bowlers")
plt.ylabel("Number of wickets")
plt.show()
```

# Who are the most economical bowlers?

```python
eco=data.groupby(["bowler"]).agg(
    total_runs=("total_runs","sum"),
    total_balls=("ball","count")
).reset_index()

eco["over"]=(eco["total_balls"]/6).astype(int)
eco["economy"]=round(eco["total_runs"]/eco["over"],1)
eco
```

```python
eco=eco[eco["over"]>5]
economy=eco.sort_values("economy",ascending=False).head(10)
plt.figure(figsize=(20,5))
lab=sns.barplot(x="bowler",y="economy",data=economy,palette="plasma",alpha=0.7)
for bar in lab.containers:
    lab.bar_label(bar)
plt.title("Economical Bowlers",fontsize=13)
plt.xlabel("Bowlers")
plt.ylabel("Economy")
plt.show()
```

```python
overs=data.groupby(["over","bowler"],as_index=False)["total_runs"].sum().sort_values("over",ascending=False)
maiden=overs[overs["total_runs"]<1]
maiden
```

```python
plt.figure(figsize=(11,5))
lab=sns.barplot(x="bowler",y="over",data=maiden,palette="dark",alpha=0.5)
for bar in lab.containers:
    lab.bar_label(bar)
plt.title("Maiden Over",fontsize=13)
plt.xlabel("Bowlers")
plt.ylabel("Overs")
plt.show()
```

# What are the top partnerships by runs?

```python
partner=data.groupby(["batter","non_striker"],as_index=False)["total_runs"].sum().sort_values("total_runs",ascending=False).head(10)
partner["partnership"]=partner["batter"]+" + "+partner["non_striker"]
lab=sns.barplot(x="partnership",y="total_runs",data=partner,palette="BrBG")
plt.xticks(rotation=90)
for bar in lab.containers:
    lab.bar_label(bar)
plt.title("Partnership",fontsize=13)
plt.xlabel("Partners")
plt.ylabel("Runs")
plt.show()
```
